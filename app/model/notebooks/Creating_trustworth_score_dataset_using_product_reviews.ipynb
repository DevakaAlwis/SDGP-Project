# Import Libraries

```python
import json
import itertools
import pandas as pd
import numpy as np
import random
```

# Load the review and product datasets

```python
reviews_df=pd.read_csv('product_reviews_for_review_analysis_test_final.csv')
reviews_df
```

### FIltering dataset based on following if conditions

```python
df = pd.DataFrame()
if((reviews_df['Sentiment_label'] == 'positive') and (review_df['trustworthy_score'] >= 55) or (review_df['trustworthy_score'] <= 100)):
    pd.concat([])
```

```python
filtered_df =reviews_df[(reviews_df['Sentiment_label'] == 'positive') & (reviews_df['trustworthy_score'] >= 55) |
                 (reviews_df['Sentiment_label'] == 'neutral') & (reviews_df['trustworthy_score'] < 55) & (reviews_df['trustworthy_score'] >= 45) |
                 (reviews_df['Sentiment_label'] == 'negative') & (reviews_df['trustworthy_score'] < 45)]
filtered_df
```

```python

a = filtered_df[filtered_df['Sentiment_label'] == 'positive'].count()
b = filtered_df[filtered_df['Sentiment_label'] == 'neutral'].count()
c = filtered_df[filtered_df['Sentiment_label'] == 'negative'].count()
a + b + c
```

```python
rating = filtered_df.groupby('asin')['overall'].mean()
rating
```

```python
review_count = filtered_df.groupby('asin')['asin'].count()
review_count
```

```python
positive_reviews = filtered_df[filtered_df['Sentiment_label'] == 'positive'].groupby('asin').size().rename('positive_count')
positive_reviews.max()
```

```python
negative_reviews = filtered_df[filtered_df['Sentiment_label'] == 'negative'].groupby('asin').size().rename('negative_count')
negative_reviews
```

```python
neutral_reviews = filtered_df[filtered_df['Sentiment_label'] == 'neutral'].groupby('asin').size().rename('neutral_count')
neutral_reviews
```

```python
trustworthy_score = filtered_df.groupby('asin')['trustworthy_score'].mean()
trustworthy_score
```

## Creating new dataframe

```python
new_df = pd.DataFrame()
```

```python
new_df['rating'] = rating
new_df['review_count'] = review_count
new_df['positive_reviews'] = positive_reviews
new_df['neutral_reviews'] = neutral_reviews
new_df['negative_reviews'] = negative_reviews
new_df['trustworthy_score'] = trustworthy_score
new_df = new_df.reset_index()
new_df
```

```python
new_df['positive_reviews'] = new_df['positive_reviews'].fillna(0)
new_df['negative_reviews'] = new_df['negative_reviews'].fillna(0)
new_df['neutral_reviews'] = new_df['neutral_reviews'].fillna(0)
new_df
```
